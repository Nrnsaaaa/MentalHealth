```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('survey.csv')
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df.drop(['comments', 'state', 'Timestamp'], axis=1, inplace=True)
```

```python
df['Gender'] = df['Gender'].str.lower().replace({
    'male': 'male', 'm': 'male', 'female': 'female', 'f': 'female'
})
```

```python
df = df.dropna(subset=['Age', 'Gender', 'self_employed', 'treatment'])
```

```python
sns.countplot(data=df, x='Gender')
plt.title('Distribusi Gender')
```

```python
df['treatment'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Apakah Pernah Menjalani Pengobatan?')
```

```python
sns.countplot(x='benefits', data=df)
plt.title('Perusahaan Menyediakan Dukungan Mental Health')
```

```python
sns.boxplot(x='treatment', y='Age', data=df)
```
